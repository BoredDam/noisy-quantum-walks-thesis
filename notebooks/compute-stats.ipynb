```python
def var_mu0(X, Y):
    X_arr = np.asarray(X)
    Y_arr = np.asarray(Y)
    return np.sum(Y_arr * (X_arr ** 2))
```

```python
def mean(X, Y):
    X = np.asarray(X)
    Y = np.asarray(Y)
    
    mean = np.sum(X * Y)
    return mean
```

```python
def skewness(X, Y):

    X = np.asarray(X)
    Y = np.asarray(Y)
    
    mean = np.sum(X * Y)
    
    variance = np.sum(((X - mean) ** 2) * Y)
    std_dev = np.sqrt(variance)
    moment_3 = np.sum(((X - mean) ** 3) * Y)
    
    return moment_3 / (std_dev ** 3)


def kurtosis(X, Y, excess=False):

    X = np.asarray(X)
    Y = np.asarray(Y)
    
    mean = np.sum(X * Y)
    variance = np.sum(((X - mean) ** 2) * Y)
    std_dev = np.sqrt(variance)

    moment_4 = np.sum(((X - mean) ** 4) * Y)
    kurtosis_pearson = moment_4 / (std_dev ** 4)
    
    if excess:
        return kurtosis_pearson - 3
    return kurtosis_pearson
```

```python
from src.random_walk.random_walk import random_walk

CY, CX = random_walk(100, 100, 0.5) # CASO CLASSICO

c_var = var_mu0(CX, CY)
c_kurt = kurtosis(CX, CY)
```

```python
import csv, os
folder_name = "noisy_walk_H"
results = []

for a in os.scandir(f"./data/{folder_name}/"):
    if a.name.endswith("mean.csv"):
        with open(a.path) as csvfile:
            p = float(a.name.removesuffix("_mean.csv").removeprefix(f"{folder_name}_p_"))
            rowreader = csv.reader(csvfile)
            data = list(rowreader)
            X = [int(i) for i in data[0]]
            Y = [float(i) for i in data[1]]
            mu = mean(X, Y)
            var = var_mu0(X, Y)
            skew = skewness(X, Y)
            kurt = kurtosis(X, Y)
            results.append((p, mu, var, skew, kurt))
```
