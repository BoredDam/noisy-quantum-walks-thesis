```python
import matplotlib.pyplot as plt
import numpy as np
from src.quantum_random_walk.quantum_random_walk import noisy_circle_quantum_random_walk_1D, noisy_meas_circle_quantum_random_walk_1D
```

```python
max_pos = 100
steps = 100
symm_qrw = True
SEED = 42
P = np.linspace(0,0.7,20)
dists = []
TIMES = 200



for p in P:
    np.random.seed(SEED)
    for t in range(TIMES):
        data, X = noisy_circle_quantum_random_walk_1D(max_pos, steps, 0.5, symm_qrw, 'H', p, None)
        Y = data[-1]
        X_odd = X[::2]
        Y_odd = Y[0::2]
        dists.append(Y)
    arr_dists = np.array(dists)
    arr_sum = np.sum(arr_dists, 0)
    Y = arr_sum / TIMES
    X_odd = X[::2]
    Y_odd = Y[0::2]
    dists.append(Y)
    plt.grid(True, which='major')
    plt.bar(X, Y, alpha=0.6)
    plt.plot(X_odd, Y_odd, '--')
    plt.title(f"p={p}")
    plt.show()
```
